## Part 1

```python
import re
import pandas as pd
```

```python
# Import the input file into a DataFrame
vent_list = []
with open("inputs/day5.txt") as input_file:
    vent_list = pd.DataFrame([list(map(int, re.findall(r"[0-9]+", x))) for x in input_file.read().splitlines()], columns=["x1", "y1", "x2", "y2"])

vent_list.head()
```

```python
# Find max value for x and y
max_x, max_y = vent_list[["x1", "x2"]].max().max() +1, vent_list[["y1", "y2"]].max().max() +1
print(max_x, max_y)
    
# Create the diagram table
vent_diagram = pd.DataFrame([[0 for x in range(max_x)] for y in range(max_y)])
vent_diagram.head(10)
```

```python
# Apply lines to diagram
for index, row in vent_list[vent_list["x1"] == vent_list["x2"]].iterrows():
    for y in range(row[["y1", "y2"]].min(), row[["y1", "y2"]].max()+1):
        vent_diagram[row["x1"]][y] += 1

for index, row in vent_list[vent_list["y1"] == vent_list["y2"]].iterrows():
    for x in range(row[["x1", "x2"]].min(), row[["x1", "x2"]].max()+1):
        vent_diagram[x][row["y1"]] += 1

vent_diagram.head(10)
```

```python
# Calculate the number of overlaps
number_of_overlap = 0
for i in range(max_x):
    number_of_overlap += vent_diagram.loc[i][vent_diagram.loc[i] >= 2].count()

print(number_of_overlap)
```

## Part 2

```python
for index, row in vent_list[(vent_list["x1"] != vent_list["x2"]) & (vent_list["y1"] != vent_list["y2"])].iterrows():  
    if row["y1"] < row["y2"]:
        start_x, stop_x = row["x1"], row["x2"]
        start_y , stop_y = row["y1"], row["y2"]
    else:
        start_x, stop_x = row["x2"], row["x1"]
        start_y , stop_y = row["y2"], row["y1"]
    if start_x < stop_x:
        step = 1
    else:
        step = -1
    for i in range(stop_y - start_y +1):
        vent_diagram[start_x+i*step][start_y+i] += 1
vent_diagram.head(10)
```

```python
# Calculate the number of overlaps
number_of_overlap = 0
for i in range(max_x):
    number_of_overlap += vent_diagram.loc[i][vent_diagram.loc[i] >= 2].count()

print(number_of_overlap)
```
